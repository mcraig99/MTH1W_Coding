```python
from matplotlib import pyplot as plt
```

```python
my_list = [1,2,3,4,5]
labels = ["My Data"]

boxprops = dict(facecolor='aquamarine', edgecolor = 'mediumslateblue', linewidth=2)
whiskerprops = dict(color='deeppink', linestyle=':', linewidth=2)
capprops = dict(color='turquoise', linewidth=2)
medianprops=dict(color='mediumslateblue', linestyle='--', linewidth=2)

plt.boxplot(my_list, vert=False, 
            patch_artist=True,          # This optional argument allows us to customize the boxplot
            
            labels=labels,              # This argument labels each data set. I have created a list of labels above, and
                                        # assigned it to the labels argument. This is helpful when you have multiple data sets.

            # The following arguments take dictionaries (assigned above), that include details like color, line style, and line
            # width for the box, whiskers, caps (ends), and median line.
            
            boxprops=boxprops,          
            whiskerprops=whiskerprops, 
            capprops=capprops,
            medianprops=medianprops)
plt.title("Look at the Colours!",fontfamily='fantasy')   # I can also use matplotlib's font families to change the font
plt.xlabel("Wow Nice",fontfamily='fantasy')
```
