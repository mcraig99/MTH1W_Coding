```python
from matplotlib import pyplot as plt
```

```python
my_list = [1,2,3,4,5]
labels = ["My Data"]

boxprops = dict(facecolor='aquamarine', edgecolor = 'mediumslateblue', linewidth=2)
whiskerprops = dict(color='deeppink', linestyle=':', linewidth=2)
capprops = dict(color='turquoise', linewidth=2)
medianprops=dict(color='mediumslateblue', linestyle='--', linewidth=2)

plt.boxplot(my_list, vert=False, 
            patch_artist=True, 
            labels=labels, 
            boxprops=boxprops, 
            whiskerprops=whiskerprops, 
            capprops=capprops,
            medianprops=medianprops)
plt.title("Look at the Colours!",fontfamily='fantasy')
plt.xlabel("Wow Nice")
```
